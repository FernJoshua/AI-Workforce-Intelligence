# Key Findings
- The dataset contains 1,470 employee records and 35 features, providing a good foundation for HR analytics.
- No missing values or duplicate records were detected, indicating excellent data quality.
- Three constant features (EmployeeCount, Over18, and StandardHours) were identified and will be removed during preprocessing.
- EmployeeNumber is a unique identifier and will be excluded from model training.
- The target variable (Attrition) is moderately imbalanced, with approximately 84% of employees remaining and 16% leaving.
- The dataset includes a rich combination of demographic, compensation, experience, performance, and satisfaction features, enabling comprehensive exploratory analysis and predictive modeling.
- Several numerical features, such as MonthlyIncome, appear to exhibit skewed distributions and will be examined further during exploratory data analysis.
- Satisfaction-related variables provide an opportunity to engineer composite features such as Satisfaction Index, Burnout Score, and Retention Risk Score.
